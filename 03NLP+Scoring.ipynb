```python
import pandas as pd
path='E:/twit_data/output2/gathered.csv'
df=pd.read_csv(path)
df
```

```python
strings=[]
for dfstring in df['text']:
    strings.append(dfstring)
strings
```

```python
import re

def remove_https_links(sentence):
    # Define the pattern to match hyperlinks starting with https
    pattern = r'https://\S+'
    # Replace the matched pattern with an empty string
    clean_sentence = re.sub(pattern, '', sentence)
    return clean_sentence
```

```python
modified_text = []

for line, string in enumerate(df['text']):
    cleared=remove_https_links(string)
    modified_text.append(cleared)
df['modified_text'] = modified_text
# Display the DataFrame
df
```

```python
df=df.drop(['withheld_in_countries','withheld_copyright','scopes'],axis=1)
df=df.drop(['quote_count','reply_count','retweet_count'],axis=1)
df=df.drop(['favorited','retweeted','source','Unnamed: 0','favorite_count'],axis=1)
df
```

# NLTK处理
SentiWordNet和词云生成需要NLTK处理

```python
import nltk
from nltk.corpus import stopwords
from nltk import ne_chunk
stop_words_English=set(stopwords.words('english'))
stop_words_add={'rt','odyssey','assassins','assassinscreedodyssey','creed','assassinscreed','ac'}
combined_stop_words = stop_words_English.union(stop_words_add)
```

```python
def _remove_stopwords(txt):
    """Delete from txt all words contained in STOPWORDS."""
    words = txt.split()
    for i, word in enumerate(words):
        if word in combined_stop_words:
            words[i] = " "
    return (" ".join(words))
```

```python
splited=[]
for string in modified_text:
    string_splited=string.split()
    splited.append(string_splited)
splited
```

```python
len(splited)
```

```python
from collections import Counter
tweets_out=[]
for tweets in splited:
    doc_out = []
    for k in tweets:        
        cleantextprep = str(k)
        # Regex cleaning
        expression = "[^a-zA-Z ]"  # keep only letters, numbers and whitespace
        cleantextCAP = re.sub(expression, '', cleantextprep)  # apply regex
        cleantext = cleantextCAP.lower()  # lower case
        cleantext = _remove_stopwords(cleantext)
        bound = ''.join(cleantext)
        doc_out.append(bound)
    tweets_out.append(doc_out)
tweets_out
```

```python
def decompose_word(doc):
    txt = []
    for word in doc:
        txt.extend(word.split())
    return txt


# decompose a list of sentences into words by self-defined function
tweets_token=[]
for tweets in tweets_out:
    tokens = decompose_word(tweets)
    tweets_token.append(tokens)
tweets_token
```

```python
len(tweets_token)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized=[]
for tweets in tweets_token:
    lemmatized_tokens=[]
    for token in tweets:
        lemmatized_token=lemmatizer.lemmatize(token)
        lemmatized_tokens.append(lemmatized_token)
    lemmatized.append(lemmatized_tokens)
lemmatized
```

```python
len(lemmatized)
```

```python
df['lemmatized_sentence'] = lemmatized
df
```

```python
def make_sentences(data,name):
    data[name]=data[name].apply(lambda x:' '.join([i+' ' for i in x]))
    # Removing double spaces if created
    data[name]=data[name].apply(lambda x:re.sub(r'\s+', ' ', x, flags=re.I))
make_sentences(df,'lemmatized_sentence')
df
```

# 进入SentiWordNet打分环节

```python

from nltk.corpus import wordnet as wn 
from nltk.corpus import sentiwordnet as swn 
postagged=[]
for lemmatized_tokens in lemmatized:
    posTagSentence = nltk.pos_tag(lemmatized_tokens) 
    postagged.append(posTagSentence)
df['post_tags']=postagged
df
```

```python
def penn_to_wn(tag):
    if tag.startswith('J'):
        return wn.ADJ
    elif tag.startswith('N'):
        return wn.NOUN
    elif tag.startswith('R'):
        return wn.ADV
    elif tag.startswith('V'):
        return wn.VERB
    return None
def get_sentiment(word,tag):
    wn_tag = penn_to_wn(tag)
    
    if wn_tag not in (wn.NOUN, wn.ADJ, wn.ADV):
        return []

    #Lemmatization
    lemma = lemmatizer.lemmatize(word, pos=wn_tag)
    if not lemma:
        return []

    #Synset is a special kind of a simple interface that is present in NLTK to look up words in WordNet. 
    #Synset instances are the groupings of synonymous words that express the same concept. 
    #Some of the words have only one Synset and some have several.
    synsets = wn.synsets(word, pos=wn_tag)
    if not synsets:
        return []

    # Take the first sense, the most common
    synset = synsets[0]
    swn_synset = swn.senti_synset(synset.name())

    return [synset.name(), swn_synset.pos_score(),swn_synset.neg_score(),swn_synset.obj_score()]

    pos=neg=obj=count=0
```

```python
senti_score = []

for pos_val in df['post_tags']:
    senti_val = [get_sentiment(x,y) for (x,y) in pos_val]
    pos=neg=0 
    for score in senti_val:
        try:
            pos = pos + score[1]  #positive score is stored at 2nd position
            neg = neg + score[2]  #negative score is stored at 3rd position
        except:
            continue
    senti_score.append(pos - neg)
     
senti_score
```

```python
from datetime import datetime
dates=[]
for time in df['created_at']:
    date=datetime.strptime(time, '%Y-%m-%d %H:%M:%S')
    dates.append(date)
df['date']=dates
```

```python
grouped_scores1 = df.groupby(df['date'].dt.date)['sentiwordscore'].apply(list).reset_index()
grouped_scores2 = df.groupby(df['date'].dt.date)['vader_compound_scores'].apply(list).reset_index()
grouped_scores= pd.merge(grouped_scores1, grouped_scores2,on='date')
grouped_scores
```

```python
summary = df.groupby(df['date'].dt.date).size().reset_index(name='count')
grouped_scores['count']=summary['count']
grouped_scores
```

```python
cs = []
for score_group in grouped_scores['sentiwordscore']:
    s = 0
    for score in score_group:
        s += score  # Assuming 'score' contains the compound score directly
    average_score = s / len(score_group)
    cs.append(average_score)
grouped_scores['average_swn_scores'] = cs

cs = []
for score_group in grouped_scores['vader_compound_scores']:
    s = 0
    for score in score_group:
        s += score  # Assuming 'score' contains the compound score directly
    average_score = s / len(score_group)
    cs.append(average_score)
grouped_scores['average_vader_scores'] = cs

grouped_scores
```
