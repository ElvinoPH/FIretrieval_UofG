```python
import pandas as pd
df = pd.read_json('D:/financial_information/twitter_data/29.json', lines=True)
df
```

```python
#将不需要的列全部清除
df=df.drop(['delete','extended_tweet','quoted_status_permalink','quoted_status'],axis=1)
df=df.drop(['quoted_status_id_str','quoted_status_id','possibly_sensitive','extended_entities'],axis=1)
df=df.drop(['display_text_range','retweeted_status','filter_level','in_reply_to_user_id_str'],axis=1)
df=df.drop(['is_quote_status','in_reply_to_screen_name','in_reply_to_user_id','in_reply_to_status_id_str'],axis=1)
df=df.drop(['geo','coordinates','contributors','id','id_str','truncated','place','in_reply_to_status_id'],axis=1)
df
```

```python
#显示这些推文中所包含的语言
unique_values = df['lang'].unique()
print(unique_values)
value_counts = df['lang'].value_counts()
print(value_counts)
```

```python
#借助数据本身的语言标签筛选出英语推文
english_twit_df = df[df['lang'].apply(lambda x: x=='en')]
english_twit_df
```
