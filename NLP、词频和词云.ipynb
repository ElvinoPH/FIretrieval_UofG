```python
dropped=[]
for string in strings:
    index = string.find("https://")
    if index != -1:  # If "https://" is found in the string
        new_string = string[:index]
        dropped.append(new_string)
dropped
```

```python
import nltk
from nltk.corpus import stopwords
from nltk import ne_chunk
nltk.download('stopwords')
stop_words_English=set(stopwords.words('english'))
```

```python
#Add the game name into the stop words
stop_words_add={'rt','odyssey','assassins','assassinscreedodyssey','creed','assassinscreed','ac'}
combined_stop_words = stop_words_English.union(stop_words_add)
```

```python
combined_stop_words
```

```python
def _remove_stopwords(txt):
    """Delete from txt all words contained in STOPWORDS."""
    words = txt.split()
    for i, word in enumerate(words):
        if word in combined_stop_words:
            words[i] = " "
    return (" ".join(words))
```

```python
splited=[]
for string in dropped:
    string_splited=string.split()
    splited.append(string_splited)
splited
```

```python
import re
from collections import Counter
tweets_out=[]
for tweets in splited:
    doc_out = []
    for k in tweets:        
        cleantextprep = str(k)
        # Regex cleaning
        expression = "[^a-zA-Z ]"  # keep only letters, numbers and whitespace
        cleantextCAP = re.sub(expression, '', cleantextprep)  # apply regex
        cleantext = cleantextCAP.lower()  # lower case
        cleantext = _remove_stopwords(cleantext)
        bound = ''.join(cleantext)
        doc_out.append(bound)
    tweets_out.append(doc_out)
tweets_out
```

```python
def decompose_word(doc):
    txt = []
    for word in doc:
        txt.extend(word.split())
    return txt


# decompose a list of sentences into words by self-defined function
tweets_token=[]
for tweets in tweets_out:
    tokens = decompose_word(tweets)
    tweets_token.append(tokens)
tweets_token
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
for tweets in tweets_token:
    for token in tweets:
        token=lemmatizer.lemmatize(token)
```

```python
flattened_list = [word for sublist in tweets_token for word in sublist]
word_freq = Counter(flattened_list)

# Print the frequency of each word
for word, freq in word_freq.items():
    print(f"{word}: {freq}")
```

```python
""" import wordcloud from python 3.7 envi """
from wordcloud import WordCloud
import matplotlib.pyplot as plt
# Join the tokenized words into a single string
text = ' '.join(flattened_list)

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud using matplotlib
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
