```python
import pandas as pd
path='E:/twit_data/output2/gathered.csv'
df=pd.read_csv(path)
df
```

```python
#移除重复和一些无关的条文
df = df.drop_duplicates(subset='text')
df = df[~df['text'].str.contains('mario', case=False)]
df.reset_index(drop=True, inplace=True)
df
```

```python
#把推文结尾的各种https://全部移除
modified_text = []

for line, string in enumerate(df['text']):
    index = string.find("https://")
    if index != -1:  # If "https://" is found in the string
        modified_text.append(string[:index])
    else:
        modified_text.append(string)
df['modified_text'] = modified_text
# Display the DataFrame
print(df['modified_text'])
```

```python
#打分
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
scores=[]
for sentence in df['modified_text']:
    vs = analyzer.polarity_scores(sentence)
    scores.append(vs)
print(len(scores))
print(len(df['modified_text']))
df['scores'] = scores
```

```python
df=df.drop(['withheld_in_countries','withheld_copyright','scopes'],axis=1)
df=df.drop(['quote_count','reply_count','retweet_count'],axis=1)
df=df.drop(['favorited','retweeted'],axis=1)
df
```

```python
import matplotlib.pyplot as plt

# Sample list of scores
compound_scores = []
for score in scores:
    compound_score=score['compound']
    compound_scores.append(compound_score)
# Plot histogram
plt.hist(compound_scores, bins=10, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Distribution of Scores')

# Show plot
plt.show()
```
