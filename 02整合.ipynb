```python
import os
import pandas as pd
path='E:/twit_data/output2'
files = os.listdir(path)
dfs=[]
json_files=[file for file in files if file.endswith('.json')]
for json_file in json_files:
    file_path = os.path.join(path,json_file)
    df = pd.read_json(file_path)
    dfs.append(df)
combined_df = pd.concat(dfs, ignore_index=True)
combined_df
```

由于还有个游戏叫什么马里奥奥德赛，得把马里奥相关的文本全部移除

```python
combined_df = combined_df.drop_duplicates(subset='text')
combined_df = combined_df[~combined_df['text'].str.contains('mario', case=False)]
combined_df.reset_index(drop=True, inplace=True)
combined_df
```

```python
combined_df.to_csv('E:/twit_data/output2/gathered.csv')
```
